```python
def parse_revenue_range(value):
    value = value.lower().replace(" ", "")
    
    if "unknown" in value or "non-appli" in value:
        return (np.nan, np.nan)
    
    elif "lessthan" in value:
        num = value.replace("lessthan", "").replace("million", "")
        return (0, float(num) * 1e6)

    elif "to" in value:
        parts = value.split("to")
        lower = parts[0]
        upper = parts[1]
        
        def convert(val):
            if "million" in val:
                return float(val.replace("million", "")) * 1e6
            elif "billion" in val:
                return float(val.replace("billion", "")) * 1e9
            elif "million" not in val and "billion" not in val:
                # Fallback: assume million
                return float(val) * 1e6
            return np.nan
        
        return (convert(lower), convert(upper))

    elif "billion" in value or "million" in value:
        # Single-point estimate like "10+billion"
        val = value.replace("+", "")
        if "billion" in val:
            num = float(val.replace("billion", "")) * 1e9
        else:
            num = float(val.replace("million", "")) * 1e6
        return (num, np.nan)

    return (np.nan, np.nan)
```

```python
df[["Revenue Lower", "Revenue Upper"]] = df["Revenue"].apply(parse_revenue_range).apply(pd.Series)
df.head()
```
